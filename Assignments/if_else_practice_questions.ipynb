```python
# 1. A student enters their marks in an exam. If the marks are greater than or
# equal to 40, print 'Pass'; otherwise print 'Fail'. Also display a message
# encouraging the student.
marks = int(input("Enter marks: "))
if marks >= 40:
    print("Pass")
    print("Great job! Keep working hard and aim higher.")
else:
    print("Fail")
    print("Don't give up. Work harder and you will succeed.")
```

```python
# 2. A user enters their age. If the age is 18 or above, allow them to vote;
# otherwise display how many years are left to become eligible.
age = int(input("Enter age: "))
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
    print("You need", 18 - age, "more years to become eligible.")
```

```python
# 13. Check whether a year entered by the user is a leap year or not and print
# an appropriate message.
year = int(input("Enter year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("It is a Leap Year.")
else:
    print("It is not a Leap Year.")
```
