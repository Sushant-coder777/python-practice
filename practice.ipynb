```python
# ans 1
li = [[1,2,3], [4,5,6], [7,8,9]]

max_value = li[0][0]
min_value = li[0][0]

for i in li:
    for j in i:
        if j > max_value:
            max_value = j
        if j < min_value:
            min_value = j 

print("max value:",max_value) 
print("min val:",min_value)  
```
