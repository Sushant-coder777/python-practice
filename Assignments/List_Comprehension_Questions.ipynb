```python
# 1. Create a list comprehension to generate squares of numbers from 1 to 10.

result = [i*i for i in range(1,11)]
print(result)
```

```python
# 2. Write a list comprehension to generate cubes of numbers from 1 to 15.

result = [i*i*i for i in range(1,16)]
print(result)
```

```python
# 3. Create a list of even numbers from 1 to 50 using list comprehension.

result = [i for i in range(1,51) if i%2==0]
print(result)
```

```python
# 4. Create a list of odd numbers from 1 to 50 using list comprehension.

result = [i for i in range(1,51) if i%2!=0]
print(result)
```
