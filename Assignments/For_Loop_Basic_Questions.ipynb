```python
# Basic for Loop and range() Usage:->

# 1. Write a program to print numbers from 1 to 10 using a for loop.

for i in range(1, 11):
    print(i)
```

```python
# 2. Print numbers from 0 to 5 using range().

for i in range(0, 6):
    print(i)
```

```python
# 3. Print numbers from 5 to 15.

for i in range(5, 16):
    print(i)
```

```python
# 4. Print numbers from 10 to 1 using a for loop.

for i in range(10, 0, -1):
    print(i)
```

```python
# 5. Print all even numbers between 1 and 20.

for i in range(2, 21, 2):
    print(i)
```

```python
# 6. Print all odd numbers between 1 and 20.

for i in range(1, 21, 2):
    print(i)
```

```python
# 7. Print numbers from 1 to 50, skipping multiples of 5.

for i in range(1, 51):
    if i % 5 != 0:
        print(i)
```

```python
# 8. Print numbers from 1 to 100 with a step of 10.

for i in range(1, 101, 10):
    print(i)
```

```python
# 9. Print the square of numbers from 1 to 10.

for i in range(1, 11):
    print(i * i)
```

```python
# 10. Print the cube of numbers from 1 to 5.

for i in range(1, 6):
    print(i ** 3)
```

```python
# Aggregation and Counting:->

# 11. Find the sum of numbers from 1 to 10 using a for loop.

total = 0
for i in range(1, 11):
    total += i
print(total)
```

```python
# 12. Find the sum of even numbers from 1 to 20.

total = 0
for i in range(2, 21, 2):
    total += i
print(total)
```

```python
# 13. Print the multiplication table of 5.

for i in range(1, 11):
    print("5 x", i, "=", 5 * i)
```

```python
# 14. Print multiplication tables from 1 to 3.

for num in range(1, 4):
    for i in range(1, 11):
        print(num, "x", i, "=", num * i)
    print()
```

```python
# 15. Count how many numbers are there between 1 and 50.

count = 0
for i in range(1, 51):
    count += 1
print(count)
```

```python
# 16. Count how many even numbers are there between 1 and 100.

count = 0
for i in range(2, 101, 2):
    count += 1
print(count)
```

```python
# Control Flow and Formatting:->

# 17. Print numbers from 1 to 30, but stop printing when the number becomes 20.

for i in range(1, 31):
    if i == 20:
        break
    print(i)
```

```python
# 18. Print numbers from 1 to 20, but do not print 10.

for i in range(1, 21):
    if i == 10:
        continue
    print(i)
```

```python
# 19. Print numbers from 1 to 5, each on a new line.

for i in range(1, 6):
    print(i)
```

```python
# 20. Print numbers from 1 to 5 in a single line separated by space.

for i in range(1, 6):
    print(i, end=" ")
```

```python
# 21. Print numbers from 5 to 1 in reverse order.

for i in range(5, 0, -1):
    print(i)
```

```python
# 22. Print numbers from 1 to 5, along with their squares.

for i in range(1, 6):
    print(i, "Square =", i * i)
```

```python
# Output Prediction and Execution Count:->

# 23. What will be the output of the following code?
for i in range(3): print(i)

for i in range(3):
    print(i)

# Output:
# 0
# 1
# 2
```

```python
# 24. What will be the output of the following code?
for i in range(1, 6, 2): print(i)

for i in range(1, 6, 2):
    print(i)

# Output:
# 1
# 3
# 5
```

```python
# 25. How many times will the following loop execute?
for i in range(10, 0, -2): print(i)

for i in range(10, 0, -2):
    print(i)

# Output:
# 10
# 8
# 6
# 4
# 2
# Loop executes 5 times
```
