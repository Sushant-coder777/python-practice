```python
# 1. Write a program to print numbers from 1 to 10 using a while loop.

i = 1
while i <= 10:
    print(i)
    i += 1
```

```python
# 12. Print the multiplication table of 7 using a while loop.

i = 1
while i <= 10:
    print("7 x", i, "=", 7 * i)
    i += 1
```

```python
# 13. Print multiplication tables from 1 to 3 using a while loop.

num = 1
while num <= 3:
    i = 1
    while i <= 10:
        print(num, "x", i, "=", num * i)
        i += 1
    print()
    num += 1
```

```python
# 14. Count how many numbers are there between 1 and 50 using a while loop.

i = 1
count = 0
while i <= 50:
    count += 1
    i += 1
print(count)
```

```python
# 16. Print numbers from 1 to 30, but stop the loop when the number becomes 20.

i = 1
while i <= 30:
    if i == 20:
        break
    print(i)
    i += 1
```

```python
# 17. Print numbers from 1 to 20, but skip printing 10.

i = 1
while i <= 20:
    if i == 10:
        i += 1
        continue
    print(i)
    i += 1
```
