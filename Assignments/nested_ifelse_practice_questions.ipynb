```python
# 1. A student enters marks. If marks are greater than or equal to 40, then
# check if marks are greater than 75 and print “Distinction”; otherwise print
# “Pass”. If marks are below 40, print “Fail”.
marks = int(input("Enter marks: "))
if marks >= 40:
    if marks > 75:
        print("Distinction")
    else:
        print("Pass")
else:
    print("Fail")
```

```python
# 17. Take a number. If it is a three-digit number, then check whether all digits
# are equal and print the result.
num = int(input("Enter number: "))
if 100 <= abs(num) <= 999:
    a = abs(num) // 100
    b = (abs(num) // 10) % 10
    c = abs(num) % 10
    if a == b == c:
        print("All digits are equal")
    else:
        print("Digits are not equal")
else:
    print("Not a three-digit number")
```

```python
# 22. A user enters password. If password length is valid, then check
# whether it contains a number and print password strength.
password = input("Enter password: ")
if len(password) >= 8:
    has_digit = any(char.isdigit() for char in password)
    if has_digit:
        print("Strong Password")
    else:
        print("Weak Password - Add at least one number")
else:
    print("Password too short")
```
