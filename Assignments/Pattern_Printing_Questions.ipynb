```python
# 1. Square Hollow Pattern

n=5
for i in range(n):
    for j in range(n):
        if i==0 or i==n-1 or j==0 or j==n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
# 2. Number Triangular Pattern

n=4
for i in range(1,n+1):
    for j in range(i):
        print(i,end=" ")
    print()
```

```python
# 3. Number Increasing Pyramid

n=4
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
# 4. Number Increasing Reverse Pyramid

n=4
for i in range(n,0,-1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
# 5. Number Changing Pyramid

n=4
num=1
for i in range(1,n+1):
    for j in range(i):
        print(num,end=" ")
        num+=1
    print()
```

```python
# 6. Zero One Triangle

n=4
for i in range(1,n+1):
    for j in range(1,i+1):
        print((i+j)%2,end=" ")
    print()
```
