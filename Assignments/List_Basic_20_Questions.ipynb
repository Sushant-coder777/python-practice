```python
# 1. Print the first element of a list

nums = [10, 20, 30, 40, 50]
print(nums[0])
```

```python
# 2. Print the last element of a list (without negative indexing)

nums = [7, 8, 9, 11]
print(nums[len(nums)-1])
```

```python
# 3. Count total number of elements in a list

nums = [1, 2, 3, 4, 5, 6]
print(len(nums))
```

```python
# 4. Check whether an element exists in the list

nums = [2, 4, 6, 8, 10]
target = 6
print(target in nums)
```

```python
# 5. Find the sum of all elements

nums = [3, 5, 7, 9, 11]
print(sum(nums))
```

```python
# 6. Find the maximum element

nums = [12, 45, 7, 89, 23]
print(max(nums))
```

```python
# 7. Find the minimum element

nums = [34, 2, 78, 5]
print(min(nums))
```

```python
# 8. Reverse a list

nums = [1, 2, 3, 4, 5]
print(nums[::-1])
```

```python
# 9. Print all elements using loop

nums = [9, 8, 7]
for i in nums:
    print(i)
```

```python
# 10. Count even numbers

nums = [1, 2, 3, 4, 5, 6]
count = 0
for i in nums:
    if i % 2 == 0:
        count += 1
print(count)
```

```python
# 11. Create list of only odd numbers

nums = [1, 2, 3, 4, 5, 6]
result = []
for i in nums:
    if i % 2 != 0:
        result.append(i)
print(result)
```

```python
# 12. Find index of an element

nums = [10, 20, 30, 40, 50]
target = 30
print(nums.index(target))
```

```python
# 13. Remove a specific element

nums = [2, 4, 6, 8, 10]
nums.remove(6)
print(nums)
```

```python
# 14. Insert element at specific position

nums = [1, 2, 4, 5]
nums.insert(2, 3)
print(nums)
```

```python
# 15. Merge two lists

list1 = [1, 2, 3]
list2 = [4, 5]
result = list1 + list2
print(result)
```

```python
# 16. Sort list in ascending order

nums = [9, 3, 7, 1, 5]
nums.sort()
print(nums)
```

```python
# 17. Find second largest element

nums = [10, 50, 30, 40, 20]
nums.sort()
print(nums[-2])
```

```python
# 18. Check list is empty or not

nums = []
if len(nums) == 0:
    print("List is Empty")
else:
    print("List is Not Empty")
```

```python
# 19. Replace negative numbers with zero

nums = [3, -2, 7, -5, 1]
for i in range(len(nums)):
    if nums[i] < 0:
        nums[i] = 0
print(nums)
```

```python
# 20. Create list of squares

nums = [2, 3, 4, 5]
result = []
for i in nums:
    result.append(i * i)
print(result)
```
